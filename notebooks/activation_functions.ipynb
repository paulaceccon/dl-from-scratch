```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def relu(z):
    return np.maximum(0, z)


def leaky_relu(z, alpha=0.01):
    return np.maximum(alpha * z, z)


def elu(z, alpha=1):
    return np.where(z < 0, alpha * (np.exp(z) - 1), z)


def derivative(f, z, eps=0.000001):
    return (f(z + eps) - f(z - eps)) / (2 * eps)
```

```python
z = np.linspace(-5, 5, 200)

plt.figure(figsize=(14, 6))

plt.subplot(121)
plt.plot(z, np.sign(z), linewidth=1, label="Step")
plt.plot(z, sigmoid(z), linewidth=2, label="Sigmoid")
plt.plot(z, np.tanh(z), linewidth=2, label="Tanh")
plt.plot(z, relu(z), "-.", linewidth=2, label="ReLU")
plt.plot(z, leaky_relu(z), ":", linewidth=2, label="Leaky ReLU")
plt.plot(z, elu(z), "--", linewidth=2, label="ELU")
plt.grid(True)
plt.legend(loc="center right", fontsize=14)
plt.title("Activation functions", fontsize=14)
plt.axis([-5, 5, -1.2, 1.2])

plt.subplot(122)
plt.plot(z, derivative(np.sign, z), linewidth=1, label="Step")
plt.plot(0, 0, "ro", markersize=5)
plt.plot(0, 0, "rx", markersize=10)
plt.plot(z, derivative(sigmoid, z), linewidth=2, label="Sigmoid")
plt.plot(z, derivative(np.tanh, z), linewidth=2, label="Tanh")
plt.plot(z, derivative(relu, z), "-.", linewidth=2, label="ReLU")
plt.plot(z, derivative(leaky_relu, z), ":", linewidth=2, label="Learky ReLU")
plt.plot(z, derivative(elu, z), "--", linewidth=2, label="ELU")
plt.grid(True)
plt.title("Derivatives", fontsize=14)
plt.axis([-5, 5, -0.2, 1.2]);
```
